```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Country_data2.csv')
df.head(25)
```

```python
dataset = pd.read_csv('Country_data2.csv')
x = dataset.iloc[:,[3,5]].values
x [:25]
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1, 11):
  kmeans = KMeans (n_clusters = i, init = 'k-means++', random_state = 42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('elbow method')
plt.xlabel('Cluster Member')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict (x)

dataset['member_cluster'] = y_kmeans + 1
dataset.head(25)
```

```python
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1], s = 30, c= 'red', label= 'cluster 1')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], s = 30, c= 'blue', label= 'cluster 2')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1], s = 30, c= 'purple', label= 'cluster 3')
plt.scatter(x[y_kmeans == 3,0], x[y_kmeans == 3,1], s = 30, c= 'green', label= 'cluster 4')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=50, c = 'yellow' , label = 'Centroids')
plt.title('clusters of Health & income The Countries')
plt.xlabel('Tingkat Kesehatan')
plt.ylabel('Tingkat Pendapatan')
plt.legend()
plt.show()
```

```python
sumbux="Esport_Wins"

plt.subplots(figsize=(6,4), dpi=100)
sns.histplot( dataset.loc[dataset.member_cluster==4,sumbux] , color="green", label="cluster 4")
sns.histplot( dataset.loc[dataset.member_cluster==3,sumbux] , color="purple", label="cluster 3")
sns.histplot( dataset.loc[dataset.member_cluster==2,sumbux] , color="blue", label="cluster 2")
sns.histplot( dataset.loc[dataset.member_cluster==1,sumbux] , color="red", label="cluster 1")

plt.title('histogram ' + sumbux)
plt.legend();
```
